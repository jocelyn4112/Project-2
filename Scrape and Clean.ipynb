```python
##scrape the table
tables = pd.read_html(url)
tables
```

```python
##making thetable into a dataframe 
df = tables[0]
df
```

```python
## Clean the data to match other set 
```

```python
## seperate out date time
df[['Month', 'Day', 'Time']]= df['Date / Time'].str.split("/", expand=True)


df['Year']= df['Time'].str[0:4]
df['Time']= df['Time'].str[5:10]
df.head(50)
```

```python
## fix Duration 
## split numbers from string

df['int']= df['Duration'].str[0:2]
df ['desc']=df['Duration'].str[3:9]


df['int']=df["int"].str.split("-", expand = True)
df.head(50)
```

```python
## convert desc in to time increments
## Find Unique labels
df["desc"].unique ()
```

```python
##Standardize labels seconds
df['desc'] = df['desc'].replace(['econds'],'sec')
df['desc'] = df['desc'].replace(['second'],'sec')
df['desc'] = df['desc'].replace([ ' secon'],'sec')
df['desc'] = df['desc'].replace([ 's'],'sec')
df['desc'] = df['desc'].replace(['econd' ],'sec')
df['desc'] = df['desc'].replace(['onds' ],'sec')
df['desc'] = df['desc'].replace(['sec' ],'sec')
df['desc'] = df['desc'].replace(['-20 se' ],'sec')
```

```python
##Standardize labels minutes
df['desc'] = df['desc'].replace(['mins'],'min')
df['desc'] = df['desc'].replace(['inutes'],'min')
df['desc'] = df['desc'].replace([' mins'],'min')
df['desc'] = df['desc'].replace(['minute'],'min')
df['desc'] = df['desc'].replace(['20 min'],'min')
df['desc'] = df['desc'].replace([' 25 mi'],'min')
df['desc'] = df['desc'].replace(['inute'],'min')
df['desc'] = df['desc'].replace([' minut'],'min')
df['desc'] = df['desc'].replace(['min'],'min')
df['desc'] = df['desc'].replace(['ninute'],'min')
df['desc'] = df['desc'].replace(['5 minu'],'min')
df['desc'] = df['desc'].replace([' Mins'],'min')
df['desc'] = df['desc'].replace(['minite'],'min')
```

```python
####Standardize labels hours
df['desc'] = df['desc'].replace(['ours'],'hour')
df['desc'] = df['desc'].replace(['our'],'hour')
df['desc'] = df['desc'].replace(['hours'],'hour')
df['desc'] = df['desc'].replace([' hours'],'hour')
```

```python
##selecting teh data that has the correct desc
L= ("sec", "min", "hour")
df=df[df.desc.isin(L)]
df["desc"].unique ()
df.head()
```

```python
# strip out non-numerical characters from the int column 
df['int'] = df['int'].str.extract('(\d+)', expand=False)
df['int'].unique()
```

```python
df[['int']] = df[['int']].apply(pd.to_numeric)

df.head()                                
```

```python
##fill in the duration seconds column 
df.loc[df['desc'] == 'sec', 'duration (seconds)'] =  (df['int'])*1
df.loc[df['desc'] == 'min', 'duration (seconds)'] = (df['int'])*60
df.loc[df['desc'] == 'hour', 'duration (seconds)'] =  (df['int'])*120

df.head()
```

```python
## rename the columns 
df.columns = ["datetime", "city", "state", "shape","Duration","comments","date posted", "Month","Day", "Time", "Year",  "int", "desc" ,  "duration (seconds)"]

## add lat and lon columns

df['longitude']=""
df["latitude"]= ""

##reorder and trim to needed columns
df1=df[["datetime", "city", "state", "shape","duration (seconds)","comments", "date posted","latitude", "longitude","Month","Day", "Time", "Year" ]]
df1.head()
```

```python
##call google api to link lat longitud to csv
```

```python
base_url = "https://maps.googleapis.com/maps/api/geocode/json"
city1= "Leavenworth"
state1 = "IN"
params = {"key": api_key}
params['address'] = f"{city1},{state1}"

output= requests.get(base_url, params=params).json()
output1= output["results"][0]["geometry"]["location"]["lng"]
print (output1)
```

```python

params = {"key": api_key}
count = 0
lat=[]
lon =[]


for index, row in df1.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    cty = row['city']
    ste = row['state']
    params['address'] = f"{cty},{ste}"

    # make request and convert to json
    try:
        lat_df = requests.get(base_url, params=params).json()
        lat.append(lat_df["results"][0]["geometry"]["location"]["lat"])
        lon.append(lat_df["results"][0]["geometry"]["location"]["lng"])
        
    except (KeyError, IndexError):
        count = count + 1
        strg=f'no key found for {index}, {cty}, {ste}, {count}'
        print(strg)
        

```

```python
df1["latitude"]= lat
df1['longitude']=lon
df1
```

```python
##output csv of table.
## this freezes teh data at this point and prevents having to run teh code over and ove if a mistake is made
df1.to_csv(r'cleaned_scrapped_ufo_data.csv')
```
